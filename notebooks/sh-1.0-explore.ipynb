```python
import nltk
from nltk.tokenize import word_tokenize
format_counts = df_original['Format'].value_counts()
print(format_counts)

# Function to clean the format string
def clean_format(format_string):
    # Tokenize the format string into words
    if type(format_string) == float:
        return ''
    tokens = word_tokenize(format_string.title())
    
    # Remove unwanted characters and spaces
    #cleaned_tokens = [token for token in tokens if token not in ['[', ']', "'", "''", "``", " "]]
    
    # Join the cleaned tokens back into a string
    cleaned_format = ''.join(token for token in tokens if token.isalnum())
    clean_format = cleaned_format.strip()
    
    return clean_format

# Apply the clean_format function to the 'Format' column
df_original['Format'] = df_original['Format'].apply(clean_format)
print(format_counts)
```

```python
def to_list(format_string):
    # Tokenize the format string into words
    if type(format_string) != str:
        return format_string
    
    lst = []
    genre = ''
    for letter in format_string:
        
        if letter == ',':
            lst.append(genre)
            genre = ''
        elif letter.isalnum() == True:
            genre += letter
        elif letter == ' ':
            genre += letter
        elif letter == ']':
            break
    output = []    
    for i in lst:
        output.append(i.strip())
    return output

df_original['Genres'] = df_original['Genres'].apply(to_list)
```

```python
def parse_lists(column):
    values_dict = {}
    for lst in column:
        for value in lst:
            if value not in values_dict:
                values_dict[value] = 1
            else:
                values_dict[value] += 1
    return values_dict

genre_counts = parse_lists(df_original['Genres'])
print(genre_counts)
```
